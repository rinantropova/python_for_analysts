## Task 1
* Изучите стоимости недвижимости
* Изучите распределение жилой квадратуры
* Изучите распределение по годам постройкикистройки


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc-house-data_1.csv', sep=',')
df.head()
```

```python
# Изучите стоимости недвижимости
min_price = df['price'].min()
max_price = df['price'].max()
plt.figure(figsize=(12, 6))

plt.hist(df['price'], bins=np.linspace(min_price, max_price, 50), edgecolor='black')
plt.title('Price distribution')
plt.xlabel('Price ($)')
plt.ylabel('Number of houses')

x_ticks = np.linspace(min_price, max_price, 6)
plt.xticks(x_ticks, [f'{x/1e6:.1f}M' for x in x_ticks])

plt.grid(axis='y', linestyle='--', alpha=0.7);
```

Summary: Большинство домов по стоимости располагаются в диапазоне между 100 тысячами и 1,5 млн долларов. 

```python
data = df.groupby('price')['id'].agg(count='count').reset_index().sort_values(by='price')
data.head()
```

```python
# Изучите распределение жилой квадратуры
plt.hist(df['sqft_living'], bins=15)
plt.title('Living sqft distribution')
plt.xlabel('Living sqft')
plt.ylabel('Number of houses');
```

Summary: Больше всего домов имеет жилую плошадь от 1500 до 2000 квадратных футов.

```python
# Изучите распределение по годам постройкикистройки
plt.hist(df['yr_built'], bins=15)
plt.title('Built by year distribution')
plt.xlabel('Built in year')
plt.ylabel('Number of houses');
```

Вывод: Большинство домов построены после 1950 года. Пик строительства пришелся на начало 2000х.

## Task 2
* Изучите распределение домов от наличия вида на набережную
* Изучите распределение этажей домов
* Изучите распределение состояния домов

```python
# Изучите распределение домов от наличия вида на набережную
data = df['waterfront'].value_counts()
plt.pie(data.values, autopct='%.1f%%', labels=data.index);
```

Вывод: Незначительное количество домов - всего 0,8%, - имеют вид на набережную.

```python
# Изучите распределение этажей домов
data = df['floors'].value_counts()
plt.pie(data.values, autopct='%.1f%%', labels=data.index);
```

```python
data = df['floors'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(data.index, data.values);
```

Вывод: Почти половина представленных домов имеет всего 1 этаж. Чаще всего встречаются одно- и двухэтажные дома. Меньше всего домов с этажностью 1,5 (что бы это ни значило в домостроительстве :D)

```python
#Изучите распределение состояния домов
data1 = df['condition'].value_counts()
plt.pie(data1.values, autopct='%.1f%%', labels=data1.index);
```

```python
data1 = df['condition'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(data1.index, data1.values);
```

Вывод: Большинство домов имеют состояние на уровне среднего, на троечку. Домов в плохом состоянии представлено очень мало. Есть небольшое количество домов в отличном состоянии - почти 8%.

```python
# Начнем с графика взаимного распределения для показателей цена и общая площадь дома
sns.jointplot(x=df['price'], y=df['sqft_lot'], kind='reg')
x_ticks = np.linspace(min_price, max_price, 6)
plt.xticks(x_ticks, [f'{x/1e6:.1f}M' for x in x_ticks])

plt.grid(axis='y', linestyle='--', alpha=0.7);
```

```python
# Зависимость слабовыраженная, я бы даже сказала, что ее нет. Рост цены не зависит от роста общей площади дома. 
```

```python
# Проверим тем же методом показатель жилой площади к цене.
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
x_ticks = np.linspace(min_price, max_price, 6)
plt.xticks(x_ticks, [f'{x/1e6:.1f}M' for x in x_ticks])

plt.grid(axis='y', linestyle='--', alpha=0.7);
```

```python
# Построим такую же диаграмму для оценки дома и цены
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['grade'], y=df['price'], whis=1.5)

plt.xlabel('Grade')
plt.ylabel('Price ($)')
plt.title('Distribution of price by grade')
plt.show();
```

```python
# Теперь построим корреляционную матрицу, и начнем с таблицы, где мы выберем всем числовые значения, а также принебрежем
# абсолютными значениями меньше 0.3
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = df[numeric_columns].corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
# На основании сформированной таблицы построим корреляционную матрицу:
plt.figure(figsize=(16, 14))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=5, cmap='coolwarm')
plt.title('Corralation matrix');
```

Возвращаясь к вопросу задачи 3, по этой матрице хорошо видно, что у цены на дома значение коррелирует больше всего с жилой площадью
и оценкой дома. Также есть влияние площади за минусом подвала (не знаю, как она правильно должна называться). И есть небольшая корреляция с количеством ванных комнат в доме. Этажи, спальни, набережная и вид из окна оказывают меньшее влияние (0.3, 0.3, 0.3, 0.4 соответственно).
