Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновации дома и годом постройки дома и сохраните в renv. Если реновации дома не было, то в новом признаке поставьте 0.

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df.head()
```

```python
stroki = df.head()
stroki
```

```python
avg = df['price'] / df['sqft_lot']
df['price_per_sq_lot'] = avg
df.head()
```

```python
df[df['yr_renovated'] == 0]
```

```python
df[df['yr_renovated'] > 0]
    
```

```python
df[['yr_renovated','yr_built']].info()
```

```python
def renovation(r_year, b_year):
    return r_year - b_year if r_year > 0 else 0
# renv = renovation(df['yr_renovated'], df['yr_built'])
df['delta_renovated'] = df.apply(lambda row: renovation(row['yr_renovated'], row['yr_built']), axis=1)
df.head()
```

```python
df['date'] = df['date'].str.replace('T000000', '')
df['date'].info()
```

```python
def extract_year(x):
    year = x[:4]
    return year

df['year'] = df['date'].apply(extract_year)
df.head()
```

```python
def extract_month(x):
    month = x[4:6]
    return month
    
df['month'] = df['date'].apply(extract_month)
df.head()
```

```python
new = df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```
